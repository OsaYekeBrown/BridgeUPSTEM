# Cloud Fraction Project

First, import the libraries needed

```python
import matplotlib.pyplot as plt
import numpy as np
import csv 
import pandas as pd
```

Use them to access the data for use in python

```python
csvFile = open("CloudFraction1_2018.csv", 'r') # 'r' = readable; read only
csvReader = csv.reader(csvFile, delimiter =',') # delimiter = ',' = separate thing by a comma 
cloud_lists = list(csvReader)
#print(cloud_lists)
```

```python
height = len(cloud_lists)
width = len(cloud_lists[0])
```

```python
print(height)
print(width)
```

```python
grid1= np.empty([height, width, 3], dtype = np.uint8) 
grid = np.empty([height, width, 3], dtype = np.uint8) 
```

```python
all_values = [] # We will fill in this empty list with all the data values as we go
#r = row, c = column
for r, row in enumerate(cloud_lists):
    for c, value in enumerate(row):
        if float(value) < 99999:
            all_values.append(float(value)) # This is where we fill in the list with all data values
        if 0.01 <= float(value) <= .14:
            grid[r, c] = [87, 86, 85]
        elif .14 <= float(value) <= 0.28:
            grid[r, c] = [94, 213, 255]
        elif 0.42 <= float(value) < 0.56:
            grid[r, c] = [133, 223, 255]
        elif .56 <= float(value) < 0.70:
            grid[r, c] = [158, 230, 255]
        elif .70 <= float(value) < .84:
            grid[r, c] = [189, 238, 255]
        elif 0.84 <= float(value) < .96:
            grid[r, c] = [230, 245, 250]
        elif .96 <= float(value) <= 1.00 :
            grid[r, c] = [255, 255, 255]
        
        else:
            grid[r, c] = [0, 0, 0,]
```

```python
print(max(all_values))
print(min(all_values))
```

```python
plt.title("Cloud Fraction in 2018 from 0% to 100%")
plt.scatter([], [], c =(254/255, 254/255, 254/255), label = "96% to 100% Cloud Cover", edgecolor = "black")
plt.scatter([], [], c =(230/255, 245/255, 250/255), label = "84% to 95%  Cloud Cover", edgecolor = "black")
plt.scatter([], [], c =(189/255, 238/255, 255/255), label = "70% to 83%  Cloud Cover", edgecolor = "black")
plt.scatter([], [], c =(158/255, 230/255, 255/255), label = "56% to 69%  Cloud Cover", edgecolor = "black")
plt.scatter([], [], c =(133/255, 223/255, 255/255), label = "42% to 55%  Cloud Cover", edgecolor = "black")
plt.scatter([], [], c =(94/255, 213/255, 255/255), label = "14% to 28%  Cloud Cover", edgecolor = "black")
plt.scatter([], [], c =(87/255, 86/255, 85/255), label = "1% to 14% Cloud Cover", edgecolor = "black")
plt.legend(bbox_to_anchor = [1.6, 1.0])
plt.imshow(grid)
plt.axis('off')
plt.show()
```
